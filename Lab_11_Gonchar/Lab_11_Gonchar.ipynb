–î–æ—Ç—Ä–∏–º—É–π—Ç–µ—Å—å –ø—ñ–¥–∫–∞–∑–æ–∫:

* –¥–æ–¥–∞–π—Ç–µ –≤–ª–∞—Å—Ç–∏–≤—ñ—Å—Ç—å, –ø—Ä–∏–∑–Ω–∞—á–µ–Ω—É –¥–ª—è –ø—ñ–¥—Ä–∞—Ö—É–Ω–∫—É –æ–ø–µ—Ä–∞—Ü—ñ–π `pop`, —ñ –¥–∞–π—Ç–µ –π–æ–º—É —ñ–º'—è, —â–æ –≥–∞—Ä–∞–Ω—Ç—É—î –π–æ–≥–æ –ø—Ä–∏—Ö–æ–≤—É–≤–∞–Ω–Ω—è;
* —ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É–π—Ç–µ –π–æ–≥–æ –Ω—É–ª–µ–º —É—Å–µ—Ä–µ–¥–∏–Ω—ñ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–∞;
* –¥–æ–¥–∞–π—Ç–µ –º–µ—Ç–æ–¥, —è–∫–∏–π –ø–æ–≤–µ—Ä—Ç–∞—î –ø–æ—Ç–æ—á–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è, –Ω–∞–¥–∞–Ω–µ –ª—ñ—á–∏–ª—å–Ω–∏–∫—É (–Ω–∞–∑–≤—ñ—Ç—å –π–æ–≥–æ `get_counter()`).
* –ø–µ—Ä–µ–≤—ñ—Ä—Ç–µ –∫–æ–¥

````python
class Stack:
    def __init__(self):
        self.__stk = []

    def push(self, val):
        self.__stk.append(val)

    def pop(self):
        val = self.__stk[-1]
        del self.__stk[-1]
        return val


class CountingStack(Stack):
    def __init__(self):
        Stack.__init__(self)
        self.__count = 0


    def get_counter(self):
        return self.__count


    def pop(self):
        val = Stack.pop(self)
        self.__count += 1
        return val



stk = CountingStack()
for i in range(100):
    stk.push(i)
    stk.pop()
print(stk.get_counter())
````

````python
# –ó–∞–≤–¥–∞–Ω–Ω—è 1

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from empty stack")

    def is_empty(self):
        return len(self.items) == 0

class CountingStack(Stack):
    def __init__(self):
        super().__init__()
        self.pop_count = 0  # –ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ pop –æ–ø–µ—Ä–∞—Ü–∏–π

    def pop(self):
        if not self.is_empty():
            self.pop_count += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –Ω–∞ –∫–∞–∂–¥—É—é –æ–ø–µ—Ä–∞—Ü–∏—é pop
            return super().pop()
        else:
            raise IndexError("pop from empty stack")

    def get_pop_count(self):
        return self.pop_count

# –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
stack = CountingStack()
for i in range(100):
    stack.push(i)
    stack.pop()

print(stack.get_pop_count())  # –û–∂–∏–¥–∞–µ–º—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç: 100
````

````python
# –ó–∞–≤–¥–∞–Ω–Ω—è 1
class Queue:
    def __init__(self):
        self.items = []

    def put(self, item):
        self.items.append(item)

    def get(self):
        if self.items:
            return self.items.pop(0)
        else:
            raise QueueError("Queue is empty")

    def size(self):
        return len(self.items)

class SuperQueue(Queue):
    def isempty(self):
        return len(self.items) == 0

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è –∫–æ–¥—É
que = SuperQueue()
que.put(1)
que.put("dog")
que.put(False)

for i in range(4):
    if not que.isempty():
        print(que.get())
    else:
        print("Queue empty")
````
